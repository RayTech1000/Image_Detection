```python
image = cv2.imread("/content/drive/MyDrive/01Training/Data/doggos.jpg")
box, label, count = cv.detect_common_objects(image)
output = draw_bbox(image, box, label, count)
plt.figure(figsize = (20,10))
plt.imshow(output)
print("There are " + str(label.count('dog')) + (" doggos in this image."))
plt.show()
```
